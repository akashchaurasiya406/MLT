```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('/content/train_data.csv')
```

```python
X_train = train_data['Age'].values.reshape(-1,1)
Y_train = train_data['Survived'].values.reshape(-1,1)
```

```python
plt.scatter(X_train,Y_train,marker ='x',c='b',label = 'trained value')
plt.xlabel("Age")
plt.ylabel("Survived")
plt.legend()
plt.show()
```

```python
test_data =pd.read_csv('/content/test_data.csv')
X_test = test_data['Age'].values.reshape(-1,1)
```

```python
m=0
b =0
L = 0.001
m = X_train.shape[0]
```

```python
 Y_pred = np.zeros(m)
for i in range(m):
  Y_pred = m*X_train +b
  # as cost functio = (1/(2*m))*(Y_train -(m*X_train +b) )**2
  D_m = (-1/m)*sum(X_train*(Y_train - Y_pred))
  D_b = (-1/m)*sum(Y_train -Y_pred)
  m = m-L*D_m
  b = b -L*D_b

print(m,b)
```

```python
plt.scatter(X_train,Y_train,marker ='x',c='b',label ='Actual Value')
plt.plot(X_train,Y_pred,c='r',label = 'Predicted value')
plt.xlabel("Age")
plt.ylabel("Survived")
plt.legend()
plt.show()
```
